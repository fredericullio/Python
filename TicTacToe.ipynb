```python
def tictactoe():
    import random
    from IPython.display import clear_output
    def board_display(board):
        clear_output()
        print("     |     |     ")
        print("  "+board[1]+"  |  "+board[2]+"  |  "+board[3]+"  ")
        print("_____|_____|_____")
        print("     |     |     ")
        print("  "+board[4]+"  |  "+board[5]+"  |  "+board[6]+"  ")
        print("_____|_____|_____")
        print("     |     |     ")
        print("  "+board[7]+"  |  "+board[8]+"  |  "+board[9]+"  ")
        print("     |     |     ")
    def win_check(board, mark):
        checklist=[]
        wins = [[1,5,9], [3,5,7], [7,8,9], [4,5,6], [1,2,3], [1,4,7], [2,5,8], [3,6,9]]
        for i, v in enumerate(board):
            if v == mark:
                checklist.append(i)
        for i in wins:
            for j in i[:3]:
                if j in checklist:
                    i.remove(j)
            if i == []:
                return True
    def player_choice(board):
        def space_check(position):
            return position == " "
        while True:
            select = int(input("Please, select a position from 1 to 9 for your marker. Press 0 to exit"))
            if 9>=select>=1:
                if space_check(board[select]):
                    return select
                else:
                    print("Sorry, this position is already taken.")
                    continue
            elif select == 0:
                break
            else:
                print('Please, enter a valid input.')
                continue
    def full_board_check(board):
        return " " not in board[1:10]
    def replay():
        choice = input("Do you want to play again? Y/N").upper()
        while True:
            if choice == "Y":
                return True
            elif choice == "N":
                print("Thnaks for playing! Come back again!")
                return False
            else: 
                print("Plese, answer Y or N")
                continue      
    def start():    
        chars = ['X','O']
        name = input('Player 1, please, choose your player\'s name')
        name2 = input('Player 2, please, choose your player\'s name')
        print(name,', please pick X or O as your marker. Press 0 to exit')
        while True:
            char = input().upper()
            if char == "O" or char == "X":
                chars.remove(char)
                char_2 = chars[0]
                print(name,"'s symbol is",char,"and",name2,"'s symbol is",char_2)
                return name, name2, char, char_2
            elif char == '0':
                print('Farewell! Come back again!')
                game = False
                break
            else:
                print('Please, enter correct value.')
                continue
    def whos_first(name, name2):
        if random.randint(1,2) == 1:
            print(name,' goes first!')
            return name
        else:
            print(name2,' goes first!')
            return name2
    while True:
        try:
            board = [' ']*10
            game = True
            name, name2, char, char_2 = start()
            current = whos_first(name, name2)
        except:
            break
        while game:
            if current == name:
                board_display(board)
                print(name,"'s turn")
                position = player_choice(board)
                try:    
                    board[position] = char
                except:
                    break
                if win_check(board, char):
                    print("Congratulations,",name,"! You've won!")
                    break
                elif full_board_check(board):
                    print("It's a tie!") 
                    break
                else:
                    current = name2
            else:
                board_display(board)
                print(name2,"'s turn")
                position2 = player_choice(board)
                try:
                    board[position2] = char_2
                except:
                    break
                if win_check(board, char_2):
                    print("Congratulations,",name2,"! You've won!")
                    break
                elif full_board_check(board):
                    print("It's a tie!")
                    break
                else:
                    current = name
        
        if not replay():
            break
```
